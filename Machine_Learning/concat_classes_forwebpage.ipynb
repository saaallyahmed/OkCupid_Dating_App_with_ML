```python
import os
import pandas as pd
import numpy as np
```

```python
# Raw data file
file_to_load = "train_test.csv"

# Read purchasing file and store into pandas data frame
df= pd.read_csv(file_to_load)
df.head()
```

```python
df0 = df.loc[df['class']== 0.0]
```

```python
df0=df0[df['class'] == 0.0].sample(n=5)
df0
```

```python
df2=df[df['class'] == 2.0].sample(n=5)
df2
```
