```python
from sklearn.manifold import TSNE
import pandas as pd
from sklearn.cluster import KMeans
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading data
file_path = Path("clustered.csv")
df = pd.read_csv(file_path)
df.head()
```

```python
df1=df.dropna()
```

```python
# Initialize t-SNE model
tsne = TSNE(3,learning_rate=35)
```

```python
# Reduce dimensions
tsne_features = tsne.fit_transform(df1)
```

```python
# Prepare to plot the dataset

# The first column of transformed features
df1['x'] = tsne_features[:,0]

# The second column of transformed features
df1['y'] = tsne_features[:,1]
```

```python
file_path1 = Path("clustered.csv")
df3 = pd.read_csv(file_path1)
df3.head()
```

```python
labels = df3['class']
```

```python
# Visualize the clusters
plt.scatter(df1['x'], df1['y'], c=labels)
plt.show()
```

```python
# Visualize the clusters with color
plt.scatter(df1['x'], df1['y'], c=labels)
plt.show()
```

```python
plt.scatter(df3['sex'], df3['age_scaled'], c=labels)
plt.show()
```
